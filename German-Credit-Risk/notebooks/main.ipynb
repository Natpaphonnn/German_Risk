# Phase 2: Feature Engineering
สร้าง feature ใหม่จากข้อมูลที่มี, จัดการค่า missing, แปลง categorical เป็นตัวเลข, และ scale ข้อมูล

## 19. Create New Features
สร้าง feature ใหม่:
- **Monthly Payment** = Credit Amount ÷ Duration → ค่าผ่อนต่อเดือน
- **Age Group** = แบ่งอายุเป็น Young / Adult / Senior
- **Credit per Age** = Credit Amount ÷ Age → ภาระหนี้ต่ออายุ

```python
# สร้าง DataFrame สำหรับ Feature Engineering
df_fe = df.copy()

# 1) Monthly Payment (ค่าผ่อนต่อเดือน)
df_fe["Monthly Payment"] = df_fe["Credit amount"] / df_fe["Duration"]

# 2) Age Group (แบ่งกลุ่มอายุ)
df_fe["Age Group"] = pd.cut(df_fe["Age"],
                            bins=[0, 25, 45, 100],
                            labels=["Young", "Adult", "Senior"])

# 3) Credit per Age (ภาระหนี้ต่ออายุ)
df_fe["Credit per Age"] = df_fe["Credit amount"] / df_fe["Age"]

print("New features created:")
df_fe[["Age", "Credit amount", "Duration", "Monthly Payment", "Age Group", "Credit per Age"]].head(10)
```

## 20. Visualize New Features vs Risk
ดูว่า feature ใหม่ที่สร้างขึ้นมีความแตกต่างระหว่าง good/bad อย่างไร

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Monthly Payment by Risk
sns.boxplot(x="Risk", y="Monthly Payment", data=df_fe, ax=axes[0], palette="Set2")
axes[0].set_title("Monthly Payment by Risk")

# Age Group vs Risk
sns.countplot(x="Age Group", hue="Risk", data=df_fe, ax=axes[1], palette="Set2")
axes[1].set_title("Age Group vs Risk")

# Credit per Age by Risk
sns.boxplot(x="Risk", y="Credit per Age", data=df_fe, ax=axes[2], palette="Set2")
axes[2].set_title("Credit per Age by Risk")

plt.tight_layout()
plt.show()
```

## 21. Handle Missing Values
เติมค่า NaN ใน Saving accounts (183) และ Checking account (394):
- ใช้ **"unknown"** เป็น category ใหม่ เนื่องจาก NaN อาจหมายถึง "ไม่มีบัญชี" ซึ่งเป็นข้อมูลที่มีความหมาย

```python
# ก่อนเติม: ดู missing values
print("Before filling:")
print(df_fe.isnull().sum())
print()

# เติม NaN ด้วย "unknown"
df_fe["Saving accounts"] = df_fe["Saving accounts"].fillna("unknown")
df_fe["Checking account"] = df_fe["Checking account"].fillna("unknown")

# หลังเติม: ยืนยันไม่มี missing แล้ว
print("After filling:")
print(df_fe.isnull().sum())
print(f"\nSaving accounts categories: {df_fe['Saving accounts'].unique()}")
print(f"Checking account categories: {df_fe['Checking account'].unique()}")
```

## 22. Verify "unknown" Category vs Risk
ดูว่ากลุ่ม "unknown" มี bad risk rate ต่างจากกลุ่มอื่นอย่างไร

```python
# Bad risk rate ของแต่ละ category (รวม unknown)
for col in ["Saving accounts", "Checking account"]:
    print(f"\n{'='*40}")
    print(f"  {col} — Bad Risk Rate")
    print(f"{'='*40}")
    ct = pd.crosstab(df_fe[col], df_fe["Risk"], normalize="index")
    ct.columns = ["good_rate", "bad_rate"]
    ct["count"] = df_fe[col].value_counts()
    print(ct.sort_values("bad_rate", ascending=False).round(3))
```

## 23. Encode Categorical Variables
แปลง categorical columns เป็นตัวเลขด้วย Label Encoding สำหรับ tree-based models
- Sex, Housing, Saving accounts, Checking account, Purpose, Age Group

```python
from sklearn.preprocessing import LabelEncoder

# Encode target
df_fe["Risk_encoded"] = df_fe["Risk"].map({"good": 0, "bad": 1})

# Encode categorical columns
cat_cols = ["Sex", "Housing", "Saving accounts", "Checking account", "Purpose", "Age Group"]
encoders = {}

for col in cat_cols:
    le = LabelEncoder()
    df_fe[f"{col}_encoded"] = le.fit_transform(df_fe[col].astype(str))
    encoders[col] = le
    print(f"{col}: {dict(zip(le.classes_, le.transform(le.classes_)))}")

print(f"\nEncoders saved: {list(encoders.keys())}")
```

## 24. Prepare Final Feature Set & Scale
เลือก features สำหรับ modeling, scale ด้วย StandardScaler, และแบ่ง train/test

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# เลือก features (ใช้ encoded columns + numeric features ใหม่)
feature_cols = [
    "Age", "Job", "Credit amount", "Duration",
    "Monthly Payment", "Credit per Age",
    "Sex_encoded", "Housing_encoded",
    "Saving accounts_encoded", "Checking account_encoded",
    "Purpose_encoded", "Age Group_encoded",
]

X = df_fe[feature_cols]
y = df_fe["Risk_encoded"]

print(f"Features: {X.shape[1]} columns")
print(f"Target distribution:\n{y.value_counts()}\n")

# Scale
scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

# Train/Test split (80/20, stratified)
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

print(f"Train: {X_train.shape[0]} rows | Test: {X_test.shape[0]} rows")
print(f"Train target ratio: {y_train.value_counts(normalize=True).round(3).to_dict()}")
print(f"Test target ratio:  {y_test.value_counts(normalize=True).round(3).to_dict()}")
```

## 25. Verify Scaled Data
ดูข้อมูลหลัง scaling (mean ≈ 0, std ≈ 1)

```python
# ตรวจสอบว่า scaling ถูกต้อง
print("Scaled data stats (should be mean≈0, std≈1):")
print(X_scaled.describe().round(2).loc[["mean", "std"]])
print(f"\nX_train shape: {X_train.shape}")
print(f"X_test shape:  {X_test.shape}")
X_scaled.head()
```

# Phase 3: Modeling & Comparison
เทรนหลายโมเดล, เปรียบเทียบผล, จัดการ class imbalance, และ tune hyperparameters

## 26. Train Baseline Models
เทรน 5 โมเดลพื้นฐาน (ไม่ tune) เพื่อดู performance เบื้องต้น:
- Logistic Regression
- Random Forest
- Gradient Boosting
- Support Vector Machine (SVM)
- XGBoost

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, roc_auc_score, classification_report

# Define baseline models
baseline_models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=200, random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(n_estimators=200, random_state=42),
    "SVM": SVC(probability=True, random_state=42),
    "XGBoost": XGBClassifier(n_estimators=200, random_state=42, eval_metric="logloss"),
}

# Train & evaluate
baseline_results = {}
for name, model in baseline_models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]

    acc = accuracy_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_prob)
    baseline_results[name] = {"accuracy": acc, "auc": auc, "model": model}

    print(f"\n{'='*50}")
    print(f"  {name}")
    print(f"{'='*50}")
    print(f"  Accuracy : {acc:.4f}")
    print(f"  ROC-AUC  : {auc:.4f}")
    print(classification_report(y_test, y_pred, target_names=["good", "bad"]))
```

## 27. Baseline Comparison Chart
เปรียบเทียบ Accuracy และ ROC-AUC ของ 5 โมเดลในกราฟเดียว

```python
# Baseline comparison bar chart
names = list(baseline_results.keys())
accs = [baseline_results[n]["accuracy"] for n in names]
aucs = [baseline_results[n]["auc"] for n in names]
colors = ["#66c2a5", "#fc8d62", "#8da0cb", "#e78ac3", "#a6d854"]

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

bars1 = axes[0].barh(names, accs, color=colors[:len(names)])
axes[0].set_xlim(0.5, 1.0)
axes[0].set_title("Accuracy")
axes[0].bar_label(bars1, fmt="%.4f", padding=3)

bars2 = axes[1].barh(names, aucs, color=colors[:len(names)])
axes[1].set_xlim(0.5, 1.0)
axes[1].set_title("ROC-AUC")
axes[1].bar_label(bars2, fmt="%.4f", padding=3)

plt.tight_layout()
plt.show()
```

## 28. Cross-Validation (5-Fold)
ใช้ 5-Fold CV เพื่อวัดความเสถียรของแต่ละโมเดล (ลด overfitting bias จากการ split ครั้งเดียว)

```python
from sklearn.model_selection import cross_val_score

cv_models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=200, random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(n_estimators=200, random_state=42),
    "SVM": SVC(probability=True, random_state=42),
    "XGBoost": XGBClassifier(n_estimators=200, random_state=42, eval_metric="logloss"),
}

cv_results = {}
for name, model in cv_models.items():
    acc_scores = cross_val_score(model, X_scaled, y, cv=5, scoring="accuracy")
    auc_scores = cross_val_score(model, X_scaled, y, cv=5, scoring="roc_auc")
    cv_results[name] = {
        "acc_mean": acc_scores.mean(), "acc_std": acc_scores.std(),
        "auc_mean": auc_scores.mean(), "auc_std": auc_scores.std(),
    }
    print(f"{name:25s} | Acc: {acc_scores.mean():.4f} ± {acc_scores.std():.4f} | AUC: {auc_scores.mean():.4f} ± {auc_scores.std():.4f}")

# Summary table
cv_df = pd.DataFrame(cv_results).T
cv_df.columns = ["Acc Mean", "Acc Std", "AUC Mean", "AUC Std"]
print("\n")
cv_df.round(4)
```

## 29. Handle Class Imbalance with SMOTE
ข้อมูล imbalanced (good 70% / bad 30%) → ใช้ SMOTE สร้างข้อมูล bad เพิ่มใน training set แล้วเทรนโมเดลใหม่เปรียบเทียบ

```python
from imblearn.over_sampling import SMOTE

# Apply SMOTE on training set only
smote = SMOTE(random_state=42)
X_train_sm, y_train_sm = smote.fit_resample(X_train, y_train)

print(f"Before SMOTE: {y_train.value_counts().to_dict()}")
print(f"After SMOTE:  {y_train_sm.value_counts().to_dict()}")

# Re-train all models with SMOTE data
smote_results = {}
for name, model in cv_models.items():
    model_sm = model.__class__(**model.get_params())
    model_sm.fit(X_train_sm, y_train_sm)
    y_pred = model_sm.predict(X_test)
    y_prob = model_sm.predict_proba(X_test)[:, 1]

    acc = accuracy_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_prob)
    smote_results[name] = {"accuracy": acc, "auc": auc, "model": model_sm}

    print(f"\n{'='*50}")
    print(f"  {name} (SMOTE)")
    print(f"{'='*50}")
    print(f"  Accuracy : {acc:.4f}")
    print(f"  ROC-AUC  : {auc:.4f}")
    print(classification_report(y_test, y_pred, target_names=["good", "bad"]))
```

## 30. Baseline vs SMOTE Comparison
เปรียบเทียบ ROC-AUC ระหว่าง Baseline (ไม่ SMOTE) กับ SMOTE

```python
# Compare Baseline vs SMOTE
compare_data = []
for name in baseline_results:
    compare_data.append({
        "Model": name,
        "Baseline Acc": baseline_results[name]["accuracy"],
        "SMOTE Acc": smote_results[name]["accuracy"],
        "Baseline AUC": baseline_results[name]["auc"],
        "SMOTE AUC": smote_results[name]["auc"],
    })

compare_df = pd.DataFrame(compare_data).set_index("Model")
print(compare_df.round(4))

# Plot comparison
fig, ax = plt.subplots(figsize=(14, 5))
x = np.arange(len(compare_df))
width = 0.35

bars1 = ax.bar(x - width/2, compare_df["Baseline AUC"], width, label="Baseline", color="#8da0cb")
bars2 = ax.bar(x + width/2, compare_df["SMOTE AUC"], width, label="SMOTE", color="#fc8d62")

ax.set_ylabel("ROC-AUC")
ax.set_title("Baseline vs SMOTE — ROC-AUC Comparison")
ax.set_xticks(x)
ax.set_xticklabels(compare_df.index, rotation=15)
ax.set_ylim(0.5, 1.0)
ax.legend()
ax.bar_label(bars1, fmt="%.3f", padding=3, fontsize=9)
ax.bar_label(bars2, fmt="%.3f", padding=3, fontsize=9)

plt.tight_layout()
plt.show()
```

## 31. Hyperparameter Tuning (GridSearchCV)
ปรับ hyperparameters ของ 3 โมเดล tree-based (Random Forest, Gradient Boosting, XGBoost) ด้วย GridSearchCV + SMOTE

```python
from sklearn.model_selection import GridSearchCV

# --- Random Forest ---
rf_params = {
    "n_estimators": [100, 200, 300],
    "max_depth": [5, 10, 15, None],
    "min_samples_split": [2, 5, 10],
}

rf_grid = GridSearchCV(
    RandomForestClassifier(random_state=42),
    rf_params, cv=5, scoring="roc_auc", n_jobs=-1, verbose=0
)
rf_grid.fit(X_train_sm, y_train_sm)

print("=== Random Forest Best Params ===")
print(f"  Params: {rf_grid.best_params_}")
print(f"  CV AUC: {rf_grid.best_score_:.4f}")

# --- Gradient Boosting ---
gb_params = {
    "n_estimators": [100, 200, 300],
    "max_depth": [3, 5, 7],
    "learning_rate": [0.05, 0.1, 0.2],
}

gb_grid = GridSearchCV(
    GradientBoostingClassifier(random_state=42),
    gb_params, cv=5, scoring="roc_auc", n_jobs=-1, verbose=0
)
gb_grid.fit(X_train_sm, y_train_sm)

print("\n=== Gradient Boosting Best Params ===")
print(f"  Params: {gb_grid.best_params_}")
print(f"  CV AUC: {gb_grid.best_score_:.4f}")

# --- XGBoost ---
xgb_params = {
    "n_estimators": [100, 200, 300],
    "max_depth": [3, 5, 7],
    "learning_rate": [0.05, 0.1, 0.2],
}

xgb_grid = GridSearchCV(
    XGBClassifier(random_state=42, eval_metric="logloss"),
    xgb_params, cv=5, scoring="roc_auc", n_jobs=-1, verbose=0
)
xgb_grid.fit(X_train_sm, y_train_sm)

print("\n=== XGBoost Best Params ===")
print(f"  Params: {xgb_grid.best_params_}")
print(f"  CV AUC: {xgb_grid.best_score_:.4f}")
```

## 32. Evaluate Tuned Models on Test Set
วัดผล tuned models บน test set ที่ไม่เคยเห็น

```python
# Evaluate tuned models on test set
tuned_models = {
    "RF Tuned": rf_grid.best_estimator_,
    "GB Tuned": gb_grid.best_estimator_,
    "XGB Tuned": xgb_grid.best_estimator_,
}

tuned_results = {}
for name, model in tuned_models.items():
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]
    acc = accuracy_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_prob)
    tuned_results[name] = {"accuracy": acc, "auc": auc, "model": model}

    print(f"\n{'='*50}")
    print(f"  {name}")
    print(f"{'='*50}")
    print(f"  Accuracy : {acc:.4f}")
    print(f"  ROC-AUC  : {auc:.4f}")
    print(classification_report(y_test, y_pred, target_names=["good", "bad"]))
```

## 33. Final Model Comparison
รวมผลทั้งหมด: Baseline, SMOTE, Tuned — เปรียบเทียบในตารางเดียว

```python
# Combine all results into one table
all_results = []

for name in baseline_results:
    all_results.append({
        "Model": f"{name} (Baseline)",
        "Accuracy": baseline_results[name]["accuracy"],
        "ROC-AUC": baseline_results[name]["auc"],
    })

for name in smote_results:
    all_results.append({
        "Model": f"{name} (SMOTE)",
        "Accuracy": smote_results[name]["accuracy"],
        "ROC-AUC": smote_results[name]["auc"],
    })

for name in tuned_results:
    all_results.append({
        "Model": f"{name} (SMOTE+Tuned)",
        "Accuracy": tuned_results[name]["accuracy"],
        "ROC-AUC": tuned_results[name]["auc"],
    })

final_df = pd.DataFrame(all_results).set_index("Model").sort_values("ROC-AUC", ascending=False)
print("=== Final Model Comparison (sorted by ROC-AUC) ===\n")
final_df.round(4)
```

## 34. Select & Save Best Model
เลือกโมเดลที่ ROC-AUC สูงสุดและบันทึกไว้ใช้ใน Phase ถัดไป

```python
import joblib

# รวม model objects ทั้งหมด
all_models = {}
for name in baseline_results:
    all_models[f"{name} (Baseline)"] = baseline_results[name]
for name in smote_results:
    all_models[f"{name} (SMOTE)"] = smote_results[name]
for name in tuned_results:
    all_models[f"{name} (SMOTE+Tuned)"] = tuned_results[name]

# หา best model by AUC
best_name = max(all_models, key=lambda k: all_models[k]["auc"])
best_model = all_models[best_name]["model"]
best_auc = all_models[best_name]["auc"]
best_acc = all_models[best_name]["accuracy"]

print(f"Best Model: {best_name}")
print(f"  Accuracy: {best_acc:.4f}")
print(f"  ROC-AUC:  {best_auc:.4f}")

# Save
output_dir = os.path.join("..", "models")
os.makedirs(output_dir, exist_ok=True)
model_path = os.path.join(output_dir, "best_model.pkl")

joblib.dump({
    "model": best_model,
    "scaler": scaler,
    "encoders": encoders,
    "feature_cols": feature_cols,
    "best_name": best_name,
    "best_auc": best_auc,
}, model_path)

print(f"\nModel saved to {model_path}")
```

## 35. Confusion Matrix — All Models
พล็อต Confusion Matrix ของโมเดลที่ดีที่สุดจากแต่ละกลุ่ม (Baseline, SMOTE, Tuned) เพื่อดูว่าโมเดลทำนายผิดตรงไหน

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# เลือกโมเดลที่จะแสดง CM
cm_models = {**baseline_results, **tuned_results}

fig, axes = plt.subplots(2, 4, figsize=(20, 9))
axes = axes.flatten()

for i, (name, res) in enumerate(cm_models.items()):
    y_pred = res["model"].predict(X_test)
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(cm, display_labels=["good", "bad"])
    disp.plot(ax=axes[i], cmap="Blues", colorbar=False)
    axes[i].set_title(f"{name}\nAcc={res['accuracy']:.3f} AUC={res['auc']:.3f}", fontsize=10)

# ซ่อน axes ที่เหลือ
for j in range(i + 1, len(axes)):
    axes[j].set_visible(False)

plt.tight_layout()
plt.show()
```

## 36. ROC Curve — All Models
พล็อตเส้น ROC ของทุกโมเดลในกราฟเดียว เพื่อเปรียบเทียบความสามารถในการแยก good/bad

```python
from sklearn.metrics import roc_curve

plt.figure(figsize=(10, 8))

# รวมโมเดลทั้งหมด
roc_models = {**baseline_results, **tuned_results}

for name, res in roc_models.items():
    y_prob = res["model"].predict_proba(X_test)[:, 1]
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    plt.plot(fpr, tpr, label=f"{name} (AUC={res['auc']:.3f})", linewidth=2)

plt.plot([0, 1], [0, 1], "k--", alpha=0.5, label="Random (AUC=0.500)")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve — All Models")
plt.legend(loc="lower right", fontsize=9)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

## 37. Precision-Recall Curve
สำคัญสำหรับ imbalanced data — ดูว่าโมเดลจับ "bad" risk ได้ดีแค่ไหนโดยไม่ทำนาย good ผิดเป็น bad มากเกินไป

```python
from sklearn.metrics import precision_recall_curve, average_precision_score

plt.figure(figsize=(10, 8))

for name, res in roc_models.items():
    y_prob = res["model"].predict_proba(X_test)[:, 1]
    precision, recall, _ = precision_recall_curve(y_test, y_prob)
    ap = average_precision_score(y_test, y_prob)
    plt.plot(recall, precision, label=f"{name} (AP={ap:.3f})", linewidth=2)

plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve — All Models")
plt.legend(loc="upper right", fontsize=9)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

## 39. SHAP Values — Model Interpretation
ใช้ SHAP อธิบายว่าแต่ละ feature มีผลต่อการทำนายอย่างไร:
- **Summary Plot** — ภาพรวมว่า feature ไหนสำคัญ และค่าสูง/ต่ำดัน prediction ไปทางไหน
- **Bar Plot** — mean absolute SHAP value ของแต่ละ feature

```python
import shap

# สร้าง SHAP explainer สำหรับ best model
explainer = shap.TreeExplainer(best_model)
shap_values = explainer.shap_values(X_test)

# Summary Plot (beeswarm)
print(f"SHAP Summary — {best_name}")
shap.summary_plot(shap_values, X_test, feature_names=feature_cols, show=True)
```

## 40. SHAP Bar Plot
แสดง mean |SHAP value| ของแต่ละ feature — ยิ่งค่าสูงยิ่งมีผลต่อ prediction มาก

```python
# SHAP Bar Plot
shap.summary_plot(shap_values, X_test, feature_names=feature_cols, plot_type="bar", show=True)
```

```python
# Waterfall plot สำหรับตัวอย่างที่ 0 ใน test set
# จัดการ shap_values ให้เป็น 1D สำหรับ class "bad" (class 1)
if isinstance(shap_values, list):
    # list of [class_0_array, class_1_array]
    sv_single = shap_values[1][0]
    base_val = explainer.expected_value[1]
elif shap_values.ndim == 3:
    # shape (n_samples, n_features, n_classes)
    sv_single = shap_values[0, :, 1]
    base_val = explainer.expected_value[1] if isinstance(explainer.expected_value, (list, np.ndarray)) else explainer.expected_value
else:
    sv_single = shap_values[0]
    base_val = explainer.expected_value

shap_explanation = shap.Explanation(
    values=sv_single,
    base_values=float(base_val),
    data=X_test.iloc[0].values,
    feature_names=feature_cols
)

print(f"Actual: {'bad' if y_test.iloc[0] == 1 else 'good'}")
print(f"Predicted: {'bad' if best_model.predict(X_test.iloc[[0]])[0] == 1 else 'good'}\n")
shap.waterfall_plot(shap_explanation)
```
